```python
import akshare as ak
import pandas as pd
import pyecharts.options as opts
from pyecharts.charts import Line

# 获取股票数据
stock_pinan = ak.stock_zh_a_hist(symbol="000001", period="daily", start_date="20201001", end_date="20241231", adjust="")

# 将日期列转换为 pandas 的日期时间类型
stock_pinan['日期'] = pd.to_datetime(stock_pinan['日期'])

# 创建一个 Line 实例
line_chart = (
    Line()
    .add_xaxis(stock_pinan['日期'].dt.strftime('%Y-%m-%d').tolist())  # 将日期格式化为字符串
    .add_yaxis("最高价", stock_pinan['最高'].tolist())
    .add_yaxis("最低价", stock_pinan['最低'].tolist())
    .set_global_opts(
        title_opts=opts.TitleOpts(title="股票最高价和最低价折线图"),
        xaxis_opts=opts.AxisOpts(name="日期"),
        yaxis_opts=opts.AxisOpts(name="价格")
    )
)

# 在 Jupyter Notebook 中渲染图表
line_chart.render_notebook()
```

```python
import akshare as ak
import pandas as pd
import pyecharts.options as opts
from pyecharts.charts import Line

# 获取股票数据
stock_any = ak.stock_zh_a_hist(symbol="300059", period="daily", start_date="20131001", end_date="20241231", adjust="")

# 将日期列转换为 pandas 的日期时间类型
# stock_any['日期'] = pd.to_datetime(stock_any['日期'])
stock_any.set_index("日期", inplace=True)

# 创建一个 Line 实例
line_chart = (
    Kline()
    .add_xaxis(stock_any.index.tolist())
    .add_yaxis("300059", stock_any[["开盘","收盘","最高","最低"]].values.tolist())
    .set_global_opts(
        datazoom_opts=[opts.DataZoomOpts(type_="inside"), opts.DataZoomOpts(type_="slider")]
    )
)

# 在 Jupyter Notebook 中渲染图表
line_chart.render_notebook()
```
